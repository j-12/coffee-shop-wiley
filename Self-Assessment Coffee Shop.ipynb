```python
choice = int(input("Would you like to get started with a coffee or some baked goods? Say 1 for coffee, or 2 for baked goods: "))
```

```python
price = 0.0
if choice == 1:
    price = 0.0
    size = input("What size of coffee would you like? You can choose small, medium or large: ")
    size = size.capitalize() ##To avoid ambiguity and still understand the choice
    typeC = input("What kind of coffee do you want, choosing between brewed, espresso or coldbrew: ")
    typeC = typeC.capitalize()
    flavour = input("What flavoring do you want, if any. Choices include none, hazelnut, vanilla or caramel: ")
    flavour = flavour.capitalize()
    
    #Calculate price
    if size == 'Small':
        price += 2
    elif size == 'Medium':
        price += 3
    else:
        price += 4
    
    if typeC == 'Espresso':
        price += 0.5
    elif typeC == 'Coldbrew':
        price += 1
    else:
        price += 0
    
    if flavour == 'None':
        price += 0
    else:
        price += 0.5

    bg = int(input("Would you like some baked goods too? Press 1 for yes or 2 for no: "))
    
    if bg == 2:
        print()
        print(f'You asked for a {size} cup of {typeC} coffee with {flavour} flavouring.')
        print()
        print('Your cup of coffee costs $', price)
        total_price = price + (0.15*price)
        print('Your total price with a 15% tip is $', round(total_price,2))
    
    else:
        options = input("Would you liked a bagel, croissant or a muffin?: ")
        options = options.capitalize()
    

        if options == 'croissant':
            price += 2.50
        else:
            price += 1.50
        
        print()
        print(f'You asked for a {size} cup of {typeC} coffee with {flavour} flavouring.')
        print(f'You also want a {options}.')
        print()
        print('Your cup of coffee and baked good costs $', price)
        total_price = price + (0.15*price)
        print('Your total price with a 15% tip is $', round(total_price,2))
        

        
        
else:
    options = input("Would you liked a bagel, croissant or a muffin?: ")
    options = options.capitalize()
    price = 0.0

    if options == 'croissant':
        price += 2.50
    else:
        price += 1.50
    
    cafe = int(input("Would you also like to have a cup of coffee? Select 1 for yes or 2 for no: "))
    
    if cafe == 2:
        print()
        print(f'You asked for a {options}.')
        print()
        print("Your price for baked good is $", price)
        total_price = price + (0.15*price)
        print('Your total price with a 15% tip is $', round(total_price,2))
    
    else:
        size = input("What size of coffee would you like? You can choose small, medium or large: ")
        size = size.capitalize()
        typeC = input("What kind of coffee do you want, choosing between brewed, espresso or coldbrew: ")
        typeC = typeC.capitalize()
        flavour = input("What flavoring do you want, if any. Choices include none, hazelnut, vanilla or caramel: ")
        flavour = flavour.capitalize()

        if size == 'Small':
            price += 2
        elif size == 'Medium':
            price += 3
        else:
            price += 4

        if typeC == 'Espresso':
            price += 0.5
        elif typeC == 'Coldbrew':
            price += 1
        else:
            price += 0

        if flavour == 'None':
            price += 0
        else:
            price += 0.5
        
        print()
        print(f'You asked for a {size} cup of {typeC} coffee with {flavour} flavouring.')
        print(f'You also want a {options}.')
        print()
        print('Your cup of coffee and baked good costs $', price)
        total_price = price + (0.15*price)
        print('Your total price with a 15% tip is $', round(total_price,2))
        
    
        


        
```
